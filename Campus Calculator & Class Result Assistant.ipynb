## Question A. The Campus Calculator

```python
while True:
    print("\n" + "-"*30)
    print("Welcome to Campus Calculator!")
    print("-"*30)
    print("1. + Add")
    print("2. - Subtract")
    print("3. x Multiply")
    print("4. / Divide")
    print("5. Exit")
    print("-"*30)

    choice = input("Select operation(1-5): ")

    if choice == "5":
        print("Thank you for using the Campus Calculator.")
        print("Goodbye.")
        break

    if choice not in ["1", "2", "3", "4"]:
        print("invalid choice, please select 1-5.")
        

    num1 = float(input("First number: "))
    num2 = float(input("second number: "))

    if choice == "1":
        answer = num1 + num2
        print(num1, "+", num2, "=", answer)

    elif choice == "2":
        answer = num1 - num2
        print(num1, "-", num2, "=", answer)

    elif choice == "3":
        answer = num1 * num2
        print(num1, "x", num2, "=", answer)

    elif choice == "4":
        if num2 == 0:
            print("Error: Cannot divide by zero")
        else:
            answer = num1 / num2
            print(num1, "/", num2, "=", answer)
            
print("\n")
```

## Question B. The Class Result Assistant

```python
print("-"*30)
print("    Class Result Assistant  ")
print("-"*30 + "\n")

while True:
    try:
        no_students = int(input("How many students are in the class? "))
        if no_students > 0:
            break
        else:
            print("please enter a number bigger than 0")
    except:
        print("Please enter a number")


class_data = {}

for i in range(no_students):
    print(f"\nStudent {i+1} of {no_students}")
    name = input("Student name: ").strip()

    scores = []
    for test in range(1, 4):
        while True:
            try:
                mark = float(input(f"Test {test} score: "))
                if 0 <= mark <=100:
                    scores.append(mark)
                    break
                else:
                    print("Score must be between 0 and 100")
            except:
                print("please enter a valid number")

    class_data[name] = scores

def calculate_average(scores):
    return sum(scores) / len(scores)

def get_grade(average):
    if average >= 70: return "A"
    if average >= 60: return "B"
    if average >= 50: return "C"
    if average >= 45: return "D"
    return "F"


print("\n" + "-"*30)
print("    Class Result")
print("-" * 30)
print("=" * 60)
print(f"{'Name':<22} {'Test 1':<9} {'Test 2':<9} {'Test 3':<9} {'Avg':<8} {'Grade'}")
print("-" * 60)

for student_name, marks in class_data.items():
    Avg = calculate_average(marks)
    grade = get_grade(Avg)
    print(f"{student_name:<22} {marks[0]:<9.0f} {marks[1]:<9.0f} {marks[2]:<9.0f} {Avg:>5.1f} {grade:>4}")
    
```
